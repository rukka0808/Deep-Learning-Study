```python
img = cv2.imread('car1.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#cv2.imshow('My Car', img) 는 코랩에서 사용불가

import matplotlib.pyplot as plt

rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
plt.imshow(rgb_img)
plt.show()
plt.imshow(gray)
plt.show()
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img)

print("width: {} pixels".format(img.shape[1]))
print("height: {} pixels".format(img.shape[0]))
print("channels: {}".format(img.shape[2]))
```

```python
X = 153 #@param {type:"slider",min:0, max:1280, step:1}
Y = 354 #@param {type:"slider",min:0, max:720, step:1}
SIZE = 100
```

```python
#Crop
croped = img[Y:Y+SIZE, X:X+SIZE]
cv2_imshow(croped)
```

```python
cv2.rectangle(img, (X+SIZE*2, Y), (X+SIZE*3, Y+SIZE), (0,255,0) , 3)
cv2.putText(img, 'HELLO WORLD', (X+SIZE*3, Y), cv2.FONT_HERSHEY_COMPLEX, 2, (255,0,0))
```

```python
cv2_imshow(img)
```

```python
cv2.imwrite('car-copy.jpg', img)
```
