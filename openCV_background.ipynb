```python
img = cv2.imread('car1.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#cv2.imshow('My Car', img) 는 코랩에서 사용불가

import matplotlib.pyplot as plt

rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
plt.imshow(rgb_img)
plt.show()
plt.imshow(gray)
plt.show()
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img)

print("width: {} pixels".format(img.shape[1]))
print("height: {} pixels".format(img.shape[0]))
print("channels: {}".format(img.shape[2]))
```

```python
X = 211 #@param {type:"slider",min:0, max:1280, step:1}
Y = 158 #@param {type:"slider",min:0, max:720, step:1}
SIZE = 100
```

```python
#Crop
croped = img[Y:Y+SIZE, X:X+SIZE]
cv2_imshow(croped)
```

```python
cv2.rectangle(img, (X+SIZE*2, Y), (X+SIZE*3, Y+SIZE), (0,255,0) , 3)
```

```python
cv2_imshow(img)
cv2.putText(img, 'HELLO WORLD', (X+SIZE*3, Y), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 0))
```

```python
move = np.float32([[1, 0, 100], [0, 1, 100]])
moved = cv2.warpAffine(img, move, (2*X, 2*Y))
cv2_imshow(moved)
```

```python
(height, width) = img.shape[:2]
center = (width // 2, height // 2)

rotate = cv2.getRotationMatrix2D(center, 90, 1.0)
rotated = cv2. warpAffine(img, rotate, (width, height))
cv2_imshow(rotated)
```

```python
ratio = 500 / width
dimension = (500, int(height*ratio))
resized = cv2.resize(img, dimension, interpolation=cv2.INTER_AREA)
cv2_imshow(resized)
```

```python
flipped = cv2.flip(resized, 1)
cv2_imshow(flipped)
```

back ground

```python
 mask = np.zeros(img.shape[:2], dtype='uint8')
```

```python
cv2.circle(mask, center, int(height/2), (255,255,255), -1)
```

```python
cv2_imshow(mask )
```

```python
masked = cv2.bitwise_and(img, img, mask=mask)
cv2_imshow(masked)
```
