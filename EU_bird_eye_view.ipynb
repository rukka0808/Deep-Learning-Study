```python
img = cv2.imread('EU_test_Moment(4).jpg')
img.shape
```

```python
pts1 = np.float32([[420,400],[370,480],[640,400],[740,480]])

# 좌표의 이동점
pts2 = np.float32([[10,10],[10,800],[800,10],[800,800]])
```

```python
cv2.circle(img, (420,400), 3, (255,0,0),-1)
cv2.circle(img, (370,480), 3, (0,255,0),-1)
cv2.circle(img, (640,400), 3, (0,0,255),-1)
cv2.circle(img, (740,480), 3, (0,0,0),-1)
```

```python
M = cv2.getPerspectiveTransform(pts1, pts2)

dst = cv2.warpPerspective(img, M, (900,900))

plt.subplot(121),plt.imshow(img),plt.title('image')
plt.subplot(122),plt.imshow(dst),plt.title('Perspective')
plt.show()
```
